# 문제1
* 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
* 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후,
* 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
import pandas as pd
import numpy as np
```

```python
basic1_data='/content/basic1.csv'
```

```python
df= pd.read_csv(basic1_data)
df.head()
```

```python
# f5 기준으로 내림차순 정렬 후 상위 10개
df = df.sort_values('f5',ascending=False)
df.head(10)
```

```python
# 최소값 찾기
min = df['f5'][:10].min()
min
# min = 91.297791
```

```python
#10개를 최소값으로 대체함
df['f5'][:10] = min
df.head(10)
```

```python
# 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

print(df[df['age']>=80]['f5'].mean())
```

* 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
data70.info()
```

```python
# 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단

std = df['age'].std()
mean = df['age'].mean()

lower = mean - std*1.5
upper = mean + std*1.5

print(lower, upper)
```

```python
#'age'컬럼의 이상치를 더하시오!
df[(df['age']>upper) | (df['age']<lower) ]['age'].sum()
```

#이상치를 찾아라(소수점 나이)
* 주어진 데이터에서 이상치(소수점 나이)를 찾고 올림, 내림, 버림(절사)했을때 3가지 모두 이상치 'age' 평균을 구한 다음 모두 더하여 출력하시오
* 데이터셋 : basic1.csv

```python
data= '/content/basic1.csv'
df = pd.read_csv(data)
df.head()
```

```python
#이상치(소수점 나이)찾기

df = df[(df['age'] - np.floor(df['age']))!=0]
```

```python
# 이상치를 포함한 데이터 올림, 내림, 버림의 평균값 

#올림
m_ceil = np.ceil(df['age']).mean()
#내림
m_floor = np.floor(df['age']).mean()
#버림
m_trunc = np.trunc(df['age']).mean()

print(m_ceil,m_floor,m_trunc)
```

```python
print(m_ceil + m_floor + m_trunc)
```
